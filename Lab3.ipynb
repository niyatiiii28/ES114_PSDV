```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Q1.1 Discrete Uniform random variable
def f(m):
  x=np.arange(-m,m)
  return 1/(2*m+1)
m=int(input())
y=f(m)
x=np.arange(-m,m)
plt.bar(x,y)
plt.title('discrete uniform PMF')
plt.xlabel('value')
plt.ylabel('probability')
plt.show()

```

```python
#Q1.2 Bernoulli random variable
def f(p,k):
  pmf=[]
  for i in k:
    if i==0:
      pmf.append(1-p)
    elif i==1:
      pmf.append(p)
    else:
      pmf.append(0)
  return pmf
p=float(input())
k=np.arange(0,2)
y=f(p,k)
plt.bar(k,y)
plt.title('bernoulli PMF')
plt.xlabel('value')
plt.ylabel('probability')
plt.show()
```

```python
#Q1.3 Binomial random variable
p=float(input())
m=int(input())
k=np.arange(0,m+1)
def f(n):
  a=1
  for i in range(1,n+1):
    a=a*i
  return a
def g(p,m,k):
  pmf=[]
  for i in k:
    x=((f(m)/(f(i)*f(m-i)))*(p**i)*((1-p)**(m-i)))
    pmf.append(x)
  return pmf
y=g(p,m,k)
plt.bar(k,y)
plt.title('binomial PMF')
plt.xlabel('number of trails')
plt.ylabel('probability')
plt.show()

```

```python
#Q1.4 poisson random variable
from math import exp
p=float(input())
m=int(input())
k=np.arange(0,m+1)
lamda=m*p
def f(n):
  a=1
  for i in range(1,n+1):
    a=a*i
  return a
def g(p,m,k):
  pmf=[]
  for i in k:
    x=(exp(-lamda)*((lamda)**i)/f(i))
    pmf.append(x)
  return pmf
y=g(p,m,k)
plt.bar(k,y)
plt.title('poisson PMF')
plt.xlabel('k')
plt.ylabel('probability')
plt.show()
```

```python
#Q1.5 Geometric random variable
p=float(input())
k=int(input())
y=[]
for i in range(0,k+1):
  prob=p*((1-p)**(i-1))
  y.append(prob)
x=np.arange(0,k+1)
plt.title('geometric Probability')
plt.xlabel('number of trails')
plt.ylabel('probability of succsess')
plt.bar(x,y)
plt.show()
```

```python
#Q2 the equivalence of Binomial and Poisson
#To plot binomial random variable and poisson random variable together
from math import exp
p=float(input())
m=int(input())
k=np.arange(0,21)
lamda=m*p
def f(n):
  a=1
  for i in range(1,n+1):
    a=a*i
  return a
def g1(p,m,k):
  pmf1=[]
  for i in k:
    x=((f(m)/(f(i)*f(m-i)))*(p**i)*((1-p)**(m-i)))
    pmf1.append(x)
  plt.bar(k-0.125,pmf1,width=0.2,label='binomial',color='green')
def g2(p,m,k):
  pmf2=[]
  for i in k:
    l=(exp(-lamda)*((lamda)**i)/f(i))
    pmf2.append(l)
  plt.bar(k+0.125,pmf2,width=0.2,label='poisson',color='blue')

g1(p,m,k)
g2(p,m,k)
plt.title('PMF')
plt.xlabel('k')
plt.ylabel('number of success')
plt.legend()
plt.show()
```

from the above graphs we can say that at the large values of M, poisson and binomial random variable are the same.
