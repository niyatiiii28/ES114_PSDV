```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
#Q1(a)
def uni(a,b):
  pdf=[]
  x=np.linspace(a-1,b+1,5000)
  for i in x:
    if i>=a and i<=b:
      k=1/(b-a)
      pdf.append(k)
    else:
      pdf.append(0)
  return (pdf,x)
def plot1():
  a=int(input())
  b=int(input())
  y,x=uni(a,b)
  plt.plot(x,y)
  plt.xlabel('values')
  plt.ylabel('pdf')
  plt.title('uniform pdf')
  plt.grid()
  plt.show()
plot1()
```

```python
# Q1(b)
def normal(a,b):
  pdf=[]
  x=np.linspace(-100,100,1000)
  mean=sum(x)/len(x)
  s=0
  for i in x:
    s=s+(i-mean)**2
  sigma=math.sqrt((s/len(x)))
  for i in x:
    k=(1.0/math.sqrt(2*math.pi*sigma**2)) * math.exp( -1 * (1.0/2.0) * ( (i - mean)/sigma )**2 )
    pdf.append(k)
  return pdf,x
def plot1():
  y,x=normal(100,-100)
  plt.plot(x,y)
  plt.xlabel('values')
  plt.ylabel('pdf')
  plt.title('normal pdf')
plot1()
plt.grid()
plt.show()
```

```python
# Q1(c)
def expo(lam,a):
  pdf=[]
  x=np.linspace(0,a,1000)
  for i in x:
    k=lam*(math.exp(-1*lam*i))
    pdf.append(k)
  plt.plot(x,pdf,label=f'lambda{lam}')
  plt.xlabel('values')
  plt.ylabel('pdf')
  plt.title('exponential pdf')
a=int(input())
for i in range(3):
  lam=float(input())
  expo(lam,a)
plt.legend()
plt.grid()
plt.show()
```

```python
#Q3(a)
def f(m):
  lst=[]
  x=np.arange(-m-1,m+1)
  for i in x:
    if i>-m and i<=m:
      k=1/(2*m+1)
      lst.append(k)
    else:
      k=0
      lst.append(k)
  z=np.array(lst)
  z=z/np.sum(z)
  s=np.cumsum(z)
  plt.step(x,s,label=f'm={m}')
  plt.title('discrete uniform cdf')
  plt.xlabel('value')
  plt.ylabel('cdf')
for i in range(2):
  m=int(input())
  f(m)
plt.grid()
plt.legend()
plt.show()
```

```python
#Q3(b)
def f(p):
  pmf=[]
  k=np.arange(-2,4,1/2**10)
  for i in k:
    if i==0:
      pmf.append(1-p)
    elif i==1:
      pmf.append(p)
    else:
      pmf.append(0)
  z=np.array(pmf)
  z=z/np.sum(z)
  s=np.cumsum(z)
  plt.plot(k,s,label=f'p={p}')
  plt.title('bernoulli PMF')
  plt.xlabel('value')
  plt.ylabel('CDF')
def plot1():
  p=float(input())
  f(p)
  plt.grid()
  plt.legend()
  plt.show()
def plot2():
  p=float(input())
  f(p)
  plt.grid()
  plt.legend()
  plt.show()
plot1()
plot2()
```

```python
#Q3(c)
def f(n):
  a=1
  for i in range(1,n+1):
    a=a*i
  return a
def g(p,m):
  pmf=[]
  k=np.arange(0,m+1)
  for i in k:
    x=((f(m)/(f(i)*f(m-i)))*(p**i)*((1-p)**(m-i)))
    pmf.append(x)
  z=np.array(pmf)
  z=z/np.sum(z)
  s=np.cumsum(z)
  plt.plot(k,s,label=f'p={p}')
  plt.title('binomial cdf')
  plt.xlabel('number of trails')
  plt.ylabel('CDF')
for i in range(2):
  m=int(input())
  p=float(input())
  g(p,m)
plt.grid()
plt.legend()
plt.show()
```

```python
#Q3(d)
def f(n):
  a=1
  for i in range(1,n+1):
    a=a*i
  return a
def g(p,m):
  pmf=[]
  k=np.arange(0,30)
  lamda=m*p
  for i in k:
    x=(math.exp(-lamda)*((lamda)**i)/f(i))
    pmf.append(x)
  z=np.array(pmf)
  z=z/np.sum(z)
  s=np.cumsum(z)
  plt.plot(k,s,label=f'p={p}, m={m}')
  plt.title('poisson CDF')
  plt.xlabel('k')
  plt.ylabel('CDF')
for i in range(2):
  m=int(input())
  p=float(input())
  g(p,m)
plt.grid()
plt.legend()
plt.show()
```
