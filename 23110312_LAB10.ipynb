```python
#Q2 find the correlation between winner rank and winner point from year 2023
def summation(col):
  sonly=0
  c=0
  for i in col:
      if not np.isnan(i):
          sonly+= i
          c+= 1
  if c== 0:
      return np.nan
  return sonly
def summation1(x,y):
  s_x_y=0
  c=0
  for i in range(len(x)):
    if not (np.isnan(x[i]))  and not (np.isnan(y[i])):
      s_x_y+=x[i]*y[i]
      c+=1
  if c==0:
    return np.nan
  return s_x_y
def summation2(x,y):
  s1=0
  s2=0
  c=0
  for i in range(len(x)):
    if not (np.isnan(x[i]) or np.isnan(y[i])):
      c+=1
      s1+=x[i]
      s2+=y[i]
  if c==0:
      return np.nan
  return s1,s2
def summation3(x,y):
  s3=0
  s4=0
  c=0
  for i in range(len(x)):
     if not (np.isnan(x[i]) or np.isnan(y[i])):
        c+=1
        s3+=x[i]**2
        s4+=y[i]**2
  if c==0:
      return np.nan
  return s3,s4
def correlation(x,y):
  s1,s2=summation2(x,y)
  s3,s4=summation3(x,y)
  cov=(summation1(x,y)/len(x)-(s1/len(x)*s2/len(x)))
  var_x=np.round(((s3/len(x))-(s1/len(x))**2)**0.5,2)
  var_y=np.round(((s4/len(x))-(s2/len(x))**2)**0.5,2)
  corr=np.round(cov/(var_x*var_y),2)
  return corr
l1=[]
y3['WRank'].fillna(0, inplace=True)
y3['WPts'].fillna(0, inplace=True)
q1=correlation(y3['WRank'],y3['WPts'])
l1.append(q1)
col1=pd.Series(['(winner rank, winner points)'])
dataf=pd.DataFrame({'COLUMN':col1,'CORRELATION COEFFICIENT':l1})
dataf
```

Q7 finding covariance matrix for Winner rank , winner points , loser rank and loser points for year 2022

```python
#Q7 finding covariance matrix for Winner rank , winner points , loser rank and loser points for year 2022
def stand_data(col):
    standardized_data = (col - col.mean()) / np.std(col)
    return standardized_data

s_d1 = stand_data(y4['WRank'].fillna(0))
s_d2 = stand_data(y4['WPts'].fillna(0))
s_d3 = stand_data(y4['LRank'].fillna(0))
s_d4 = stand_data(y4['LPts'].fillna(0))

data1 = {'Winner rank': s_d1, 'Winner pts': s_d2, 'Loser rank': s_d3, 'Loser pts': s_d4}
df = pd.DataFrame(data1)
cov_matrix = df.cov()
print(cov_matrix)
```

Q8 finding eigenvalues and eigenvectors for Winner rank , winner points , loser rank and loser points for year 2022

```python
#Q8 finding eigenvalues and eigenvectors for Winner rank , winner points , loser rank and loser points for year 2022
def stand_data(col):
    standardized_data = (col - col.mean()) / np.std(col)
    return standardized_data
l=[]
s_d1 = stand_data(y4['MaxW'].fillna(0))
s_d2 = stand_data(y4['MaxL'].fillna(0))
s_d3 = stand_data(y4['AvgW'].fillna(0))
s_d4 = stand_data(y4['AvgL'].fillna(0))
l.append(s_d1)
l.append(s_d2)
l.append(s_d3)
l.append(s_d4)
data1 = {'MAxW': s_d1, 'MaxL': s_d2, 'AvgW': s_d3, 'AvgL': s_d4}
df = pd.DataFrame(data1)
cov_matrix = df.cov()

eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
 # here the eigenvector will be as column matrix 4*4 matrix
eigenvectors=np.transpose(eigenvectors)
# here it will be as row matrix
data2=['MaxW','MaxL','AvgW','AvgL']
dataf3=pd.DataFrame({'eigen Values': eigenvalues,'Eigen Vectors': np.round(eigenvectors,4).tolist()},index=data2)
print(dataf3)
print('The 4 principle components :', np.round(eigenvalues,4).tolist())
```
